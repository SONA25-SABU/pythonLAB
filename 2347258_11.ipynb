Q2. Perform the Exploratory Data Analysis on your domain-based dataset and demonstrate the retrieved insights using “Matplotlib” modules. Visualize hidden insights using appropriate plots (graphs) [Usage of line plot and scatter plot are mandatory]

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

csv_file = 'APY.csv'
df = pd.read_csv(csv_file)

# Line Graph
plt.figure(figsize=(12, 6))
plt.plot(df['Crop_Year'], df['Yield'], marker='o')
plt.title('Line Graph')
plt.xlabel('Crop_Year')

plt.ylabel('Yield')
plt.grid(True)


plt.show()
```

```python
# Bar Graph
plt.figure(figsize=(10, 6))
plt.bar(df['Crop_Year'], df['Yield'])
plt.title('Bar Graph')
plt.xlabel('Crop_Year')
plt.ylabel('Yield')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Crop_Year'], df['Yield'], c='b', marker='o')
plt.title('Scatter Plot')
plt.xlabel('Crop_Year')
plt.ylabel('Yield')
plt.grid(True)
plt.show()
```
